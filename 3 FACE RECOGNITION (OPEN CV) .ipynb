```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
```

```python
from tensorflow.keras.models import load_model
Model=load_model('Face_Recognition.h5')
```

```python
Face_Cascade=cv2.CascadeClassifier('Haarcascades/haarcascade_frontalface_default.xml')
```

```python
def Detect_Face(Image):
    Face_Image=Image.copy()
    Face_Rect=Face_Cascade.detectMultiScale(Face_Image,scaleFactor=1.2,minNeighbors=5)
    
    for (x,y,w,h) in Face_Rect:
        cv2.rectangle(Face_Image,(x,y),(x+w,y+h),(255,255,255),5)
    return Face_Image
```

```python
Video=cv2.VideoCapture(0)

while True:
    
    Ret,Frame=Video.read()
    
    if Detect_Face(Frame).all:
        
        Face=Detect_Face(Frame)
        Face_Image=cv2.resize(Face,(224,224))
        Face_Image=image.img_to_array(Face_Image)
        Face_Image=np.expand_dims(Face_Image,axis=0)  
        Pred=Model.predict(Face_Image)
        print(Pred) 
         
        Name='None'
        
        
        if(Pred[0]==0):
                Name='Biswa'
                cv2.putText(Face,Name,(50,50),cv2.FONT_HERSHEY_COMPLEX,1,(255,255,0),1)
                
        if(Pred[0]==1):
                Name='Satyam'
                cv2.putText(Face,Name,(50,50),cv2.FONT_HERSHEY_COMPLEX,1,(255,0,0),1)

    else:
        Name='None'
        cv2.putText(Face,Name,(50,50),cv2.FONT_HERSHEY_COMPLEX,1,(255,125,0),1)

                
       
    cv2.imshow('Face Recognition',Face)
        
    K=cv2.waitKey(1)
    if K==27:
        break
    
Video.release()
cv2.destroyAllWindows()
```
